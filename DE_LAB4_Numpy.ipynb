**Vectors and Matrices for Data Analysis**
**(using Numpy python Library)**

```python
#1. Calculating DISPLACEMENT

'''A person walks from origin 0 (0,0) to point A (1,2) to point B(2,1).
Calculate the distance and displacement between location points  O and B'''

import numpy as np
import cmath

O = np.array([0,0]) # position vector for location O
A = np.array([2,1]) # position vector for location A
B = np.array([3,3]) # position vector for location B

OA = A-O   # distance vector between O and A
AB = B-A   # distance vector between A and B

OAmag=np.linalg.norm(OA) # length of vector OA
print("Distance travelled from O to A =", OAmag)
ABmag=np.linalg.norm(AB) # length of vector AB
print("Distance travelled from A to B =",ABmag)

OB=OA+AB  # Displacement vector from O to B
print("Net Displacement= ",OB)

OBmag=np.linalg.norm(OB) # length of vector OA
print("Magnitude of Displacement:",OBmag)

OBangle_RAD=np.angle(complex(OB[0],OB[1])) # calculating phase angle = tan-1(y/x)=  tan-1(OB[1]/OB[0])
OBangle_DEG=np.degrees(OBangle_RAD)
print("Angle made by Displacement Vector w.r.t x axis:",OBangle_DEG)
```

```python
#2 To calculate of DOT PRODUCT (SCALAR PRODUCT) of TWO VECTORS

''' Calculate the dot product for the given two vectors
                    A = 1+1j
                    B = 2+3j
    A*B = |A||B|cos(theta) '''

VectorA = [1,1]  # Representing vector A as a 1x2 array
VectorB = [2,3]  # Representing vector B as a 1x2 array

MagA=np.linalg.norm(VectorA)
print("Magnitude of A = ",MagA)

MagB=np.linalg.norm(VectorB)
print("Magnitude of B = ",MagB)

AdotB= np.dot(VectorA,VectorB) # Determining the DOT product (A * B)using the dot function

COSangleAB=AdotB/(MagA*MagB) # cos(thetha) = A*B/|A||B|

AngleAB_Radians= np.arccos(COSangleAB) # Angle between A and B = cos-1( A*B/|A||B|)
AngleAB_Degrees=np.degrees(AngleAB_Radians)# Coverting angle in degrees

print("DOT product of vectors A and B = ", AdotB)
print("Angle between vectors A and B = ", AngleAB_Degrees)
```

```python
#3 To calculate of CROSS PRODUCT (VECTOR PRODUCT) of TWO VECTORS

''' Calculate the CROSS product for the given two vectors
                    A = 1+1j
                    B = 2+3j
    AxB = |A||B|sin(theta) '''

VectorA = np.array([1,1])  # Representing vector A as a 1x2 array
VectorB = np.array([2,3])  # Representing vector B as a 1x2 array

MagA=np.linalg.norm(VectorA)
print("Magnitude of A = ",MagA)

MagB=np.linalg.norm(VectorB)
print("Magnitude of B = ",MagB)

AcrossB= np.array([VectorA[0]*VectorB[1],-VectorA[1]*VectorB[0]])
print("CROSS product of vectors A and B = ", AcrossB)
```

```python
#4 Addition of two vectors A and B

import math

VectorA = np.array([1,1])  # Representing vector A as a 1x2 array
VectorB = np.array([2,3])  # Representing vector B as a 1x2 array

MagA=np.linalg.norm(VectorA)
print("Magnitude of A = ",MagA)

MagB=np.linalg.norm(VectorB)
print("Magnitude of B = ",MagB)

AdotB= np.dot(VectorA,VectorB) # Determining the DOT product (A * B)using the dot function

COSangleAB=AdotB/(MagA*MagB) # cos(thetha) = A*B/|A||B|

AngleAB_Radians= np.arccos(COSangleAB) # Angle between A and B = cos-1( A*B/|A||B|)

A2=MagA**2
B2=MagB**2
COSAB=math.cos(AngleAB_Radians)
SINAB=math.sin(AngleAB_Radians)
ADDmag=math.sqrt(A2+B2+(2*MagA*MagB*COSAB))
ADDangle=np.arctan((MagA*SINAB)/(MagA+(MagB*COSAB)))
ADDangle_Degrees=np.degrees(ADDangle)
print("Magnitude of A+B = ",ADDmag)
print("Angle of A+B w.r.t. A",ADDangle_Degrees)
```

```python
#7 Combining the two arrays column wise using stack

import numpy as np

arr1 = np.array([1, 2, 3])
print("Array1\n", arr1)

arr2 = np.array([4, 5, 6])
print("Array2\n", arr2)

print("Combined Array (STACKED COLUMN-WISE)\n")
arr = np.stack((arr1, arr2), axis=1)

print(arr)
```

```python
#8 Combining the two arrays ROW wise using stack

import numpy as np

arr1 = np.array([1, 2, 3])
print("Array1\n", arr1)

arr2 = np.array([4, 5, 6])
print("Array2\n", arr2)

print("Combined Array (STACKED ROW-WISE)\n")
arr = np.stack((arr1, arr2), axis=0)

print(arr)
```

```python
#9 Combining(STACKING) two arrays HORIZONTALLY (ROW-WISE) using hstack

import numpy as np

arr1 = np.array([1, 2, 3])
print("Array1\n", arr1)

arr2 = np.array([4, 5, 6])
print("Array2\n", arr2)

print("Combined Array (STACKED HORIZONTALLY)\n")
arr = np.hstack((arr1, arr2))

print(arr)
```

```python
#10 Combining(STACKING) two arrays VERTICALLY using vstack

import numpy as np

arr1 = np.array([1, 2, 3])
print("Array1\n", arr1)

arr2 = np.array([4, 5, 6])
print("Array2\n", arr2)

print("Combined Array (STACKED VERTICALLY)\n")
arr = np.vstack((arr1, arr2))

print(arr)
```

```python
# Inverse and Invertibility of a Matrix

''' A square matrix 'A' of size (nxn) is said to be INVERTIBLE,
if there exists an inverse matrix B such that
        AB= I
 where, B = Inverse(A)= A
        I = Identity matrix

Find Inverse of A =[[4,3],[3,2]]'''

A = np.array([[4,3],[3,2]])
print("Original Matrix\n",A)

A_Inverse= np.linalg.inv(A)
print("Original Matrix\n",A_Inverse)
```

```python
# Solving a system of linear equations

''' 4x-2y =  20
   -5x-5y = -10 '''

A = np.array([[4,-2],[-5,-5]])
B = np.array([20,-10])

Ainverse= np.linalg.inv(A)
X = np.dot(Ainverse,B)

print(X)
```
