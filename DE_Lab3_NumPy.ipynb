```python
# To determine DETERMINANT of a matrix

import numpy as np

arr = np.array([[1, 2],[4, 5]])
print("The given matrix is\n", arr)

DET=np.linalg.det(arr)

print("DETERMINANT of the matrix is ",DET)
```

```python
# To determine RANK of a matrix

import numpy as np

arr = np.array([[1, 2, 1],[3, 4, 7],[3, 6, 3]])
print("The given matrix is\n", arr)

RANK=np.linalg.matrix_rank(arr)

print("RANK of the matrix is ",RANK)
```

```python
# To determine Nullity of a matrix

# using RANK-Nullity Theorem
# Total No. of columns = RANK + Nullity
# EXAMPLE 1

import numpy as np

arr = np.array([[1, 2, 1],[3, 4, 7],[3, 6, 3]])
print("The given matrix is\n", arr)

RANK=np.linalg.matrix_rank(arr)  # Rank of 2-D matrix, 'arr'

COLUMNs=arr.shape[1] # No. of columns in 2-D matrix, 'arr'

print("Rank of the given matrix is ", RANK)
print("Nullity of the given matrix is ", COLUMNs-RANK)
```

```python
# To ADD and SUBTRACT two matrices

import numpy as np

arr1 = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
arr2 = np.array([[2, 4, 6, 8], [1, 1, 6, 3]])

print("MATRIX 1 + MATRIX 2\n",arr1+arr2)
print("MATRIX 1 - MATRIX 2\n",arr1-arr2)
```

```python
# To multiply two matrices using dot() function

import numpy as np

arr1 = np.array([[1, 2], [5, 6]]) # Matrix 1
arr2 = np.array([[2, 4], [1, 1]]) # Matrix 2

arr3=np.dot(arr1,arr2)

print('Multiplication of Matrix 1 and Matrix 2')
print(arr3)
```

```python
# To multiply a vector (1D array) with a matrix using dot() and transpose() function

arr1 = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
arr2 = np.array([[2, 4, 6, 8]])
arr2t=arr2.transpose()
arr3=np.dot(arr1,arr2t)
print(arr3)
```
